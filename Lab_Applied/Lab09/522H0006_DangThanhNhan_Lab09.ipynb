```python
#Exercise02
import csv
import matplotlib.pyplot as plt
import pandas as pd
# from google.colab import files
# uploaded = files.upload()

data = pd.read_csv('company-sales_data.csv')
# 1. Plot total profit
plt.plot(data['month_number'], data['total_profit'])
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.show()

# 2. Scatter plot of toothpaste
fig=plt.figure(figsize=(5, 5))
ax=fig.add_axes([0,0,1,1])
ax.scatter(data['month_number'], data['toothpaste'], color='r')
ax.set_xlabel('Month')
ax.set_ylabel('Toothpaste Sales')
ax.set_xticks(data['month_number'])
ax.set_title('Toothpaste Sales per Month')
plt.show()

# 3. Bar chart of facecream vs facewash
plt.bar(data['month_number'], data['facecream'], label='Face Cream')
plt.bar(data['month_number'], data['facewash'], bottom=data['facecream'], label='Face Wash')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#Exercise03
import re
import matplotlib.pyplot as plt
# from google.colab import files
# uploaded = files.upload()

# Read in text file
with open('Text.txt') as f:
    text = f.read()

# Preprocess text
text = text.lower()
text = re.sub(r'[^a-zA-Z\s]', '', text)
words = text.split()

# Calculate frequencies
freqs = {}
for word in words:
    if word in freqs:
        freqs[word] += 1
    else:
        freqs[word] = 1

# Get most common words
common_words = sorted(freqs, key=freqs.get, reverse=True)[:30]
counts = [freqs[w] for w in common_words]

# Plot most common words
plt.plot(common_words, counts)
plt.title('30 Most Common Words')
plt.xticks(rotation=65)
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()
```
